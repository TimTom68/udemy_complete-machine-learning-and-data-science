## Matplotlib example workflow

```python
# Import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
```

## Making figures with NumPy arrays

```python
import numpy as np
```

```python
nut_butter_prices = {'Almond': 10,
                    'Peanut': 8,
                    'Cashew': 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax.set(title='Nut butter prices',
      ylabel='Price ($)',
      xlabel='Type of butter');
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Histogram
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

# Two options for subplots

```python
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(10,5))
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5))
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

## OO method

```python
over_50 = heart_disease[heart_disease['age']>50]
over_50.head()
```

```python
#pyplot metrod
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target');
```

```python
# oo method mixed with pyplot method
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
#ax.set_xlim([45,100])
```

```python
plt.style.use('seaborn-whitegrid')
# oo method from scratch
fix, ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'],
                    cmap='summer')
ax.set(title='Heart disease and cholesterol levels',
      xlabel='Age',
      ylabel='Cholesterol');

ax.legend(*scatter.legend_elements(), title='Target');

ax.axhline(over_50['chol'].mean(),
          linestyle='--')
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10),
                               sharex=True);

# Add data to ax0
scatter = ax0.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target']);

# Customize ax0
ax0.set(title='Heart disease chart',
       xlabel='Age',
       ylabel='Cholesterol');

# Add legend
ax0.legend(*scatter.legend_elements(), title='Target')
# Add a meanline
ax0.axhline(y=over_50['chol'].mean(),
           linestyle='dotted');

# Add data to ax1
scatter2 = ax1.scatter(x=over_50['age'],
                     y=over_50['thalach'],
                     c=over_50['target']);

# Customize ax1
ax1.set(title='Heart disease chart',
       xlabel='Age',
       ylabel='Thalach');

# Add legend
ax1.legend(*scatter.legend_elements(), title='Target')
# Add a meanline
ax1.axhline(y=over_50['thalach'].mean(),
           linestyle='dashed');

fig.suptitle('Heart disease analysis', fontsize=16, fontweight='bold')
```
